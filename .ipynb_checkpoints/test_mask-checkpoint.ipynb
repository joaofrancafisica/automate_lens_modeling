```python
# Import PyAutoLens module
import autolens as al
import autolens.plot as aplt
import autofit as af
import numpy as np
from astropy.io import fits
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
img_path = './lens_light_subtraction/706/'
sim_path = './simulations/fits_files/i/705.fits'

pixel_scales = 0.263
rms_background = 19.321
exposure_time = 90.
```

```python
psf = al.Kernel2D.from_fits(file_path=img_path+'psf.fits', hdu=0, pixel_scales=pixel_scales)
# In PyAutoLens, we have to set our psf to an odd kernel or shape. In this case a 17x17 one
psf = al.preprocess.array_with_new_shape(array=psf, new_shape=(17, 17))

residual_image = al.Array2D.from_fits(file_path=sim_path, pixel_scales=pixel_scales, hdu=0)

noise_map = np.sqrt(residual_image*exposure_time+rms_background)/exposure_time
```

```python
imaging = al.Imaging(image=residual_image, noise_map=noise_map, psf=psf)
```

```python
visuals_2d = aplt.Visuals2D()
array_plotter = aplt.Array2DPlotter(array=imaging.image, visuals_2d=visuals_2d)
array_plotter.figure_2d()
```

```python
mask=al.Mask2D.from_fits(img_path+'mask.fits', pixel_scales=pixel_scales, resized_mask_shape=(100, 100))
```

```python
circ_mask = al.Mask2D.circular(shape_native=(100, 100), pixel_scales=pixel_scales, radius=10.)
```
