```python
%matplotlib inline

import autolens as al
import autolens.plot as aplt
```

```python
grid = al.Grid2DIterate.uniform(
    shape_native=(100, 100), pixel_scales=0.2633, fractional_accuracy=0.9999
)
```

```python
psf = al.Kernel2D.from_gaussian(
    shape_native=(21, 21), sigma=0.05, pixel_scales=grid.pixel_scales, normalize=True
)
```

```python
simulator = al.SimulatorImaging(
    exposure_time=2000.0, psf=psf, background_sky_level=1.0, add_poisson_noise=True
)
```

```python
lens_galaxy = al.Galaxy(redshift=0.5,
                        bulge=al.lp.EllSersic(centre=(0., 0.),
                                              elliptical_comps=al.convert.elliptical_comps_from(axis_ratio=0.9, angle=60.0),
                                              intensity=.3,
                                              effective_radius=6.,
                                              sersic_index=4),
                        mass=al.mp.EllIsothermal(centre=(0.0, 0.0),
                                                 einstein_radius=5.,
                                                 elliptical_comps=al.convert.elliptical_comps_from(axis_ratio=0.9, angle=45.0)))

source_galaxy = al.Galaxy(
    redshift=1.0,
    bulge=al.lp.EllSersic(
        centre=(0.1, -0.01),
        elliptical_comps=al.convert.elliptical_comps_from(axis_ratio=0.1, angle=30.0),
        intensity=0.3,
        effective_radius=6.,
        sersic_index=2.,
    ),
)
```

```python
tracer = al.Tracer.from_galaxies(galaxies=[lens_galaxy, source_galaxy])
```

```python
tracer_plotter = aplt.TracerPlotter(tracer=tracer, grid=grid)
tracer_plotter.figures_2d(image=True)
```
